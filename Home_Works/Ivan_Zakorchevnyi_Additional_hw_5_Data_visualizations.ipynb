`Опціональне завдання`: Пограйтесь з налаштуваннями і пошукайте варіант, за якого візуалізація розподілу суми денних платежів через seaborn є на ваш погляд зручною для аналізу.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Зробив копію даних з заміною NaN на 0
df_plot = borough_pivot.fillna(0).copy()

plt.figure(figsize=(6, 4))

# Наприклад, можна побудувати гістограми накладені
for borough in df_plot.columns:
    sns.histplot(
        df_plot[borough],
        bins=6,
        kde=False,
        stat="density",
        element="step",
        fill=True,
        alpha=0.4,
        label=borough
    )

plt.title("Розподіл сум денних платежів по районах")
plt.xlabel("Сума денних платежів")
plt.ylabel("Щільність")
plt.legend(title="Район")
plt.tight_layout()
plt.show()
```

10.З допомогою seaborn виведіть гістограми суми щомісячних платежів по кольорам автомобілів (колонка `color`). На одній фігурі має бути кількість гістограм рівна кількості різних кольорів автомобілів в наборі даних. Встановіть розмір фігури (10,5) та знайдіть значення dots-per-inch для фігури за якого якість зображення буде хорошою (зображення не буде розмитим).
`Підказки:`
1. побудувати візуалізацію можна в один рядок коду, і ще один рядок - для встановлення розмірів і роздільної здатності фігури.
2. для зміни порядку відображення гістограм кольорів автомобілів (аби менша гістограма була перед, а не за більшою) можна посортувати датафрейм за полем кольору так, аби отримати той порядок, який вам потрібен.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Зробив копію даних із заповненням NaN, щоб уникнути помилок
df_colors = df_taxi_clean.fillna('NA').copy()

# Згрупував дані по кольору, щоб можна було сортувати за сумою total
color_order = df_colors.groupby('color')['total'].sum().sort_values().index

# Встановив розмір фігури та dpi для гарної якості
plt.figure(figsize=(10,5), dpi=120)  # dpi=120 дає гарну якість для відображення на екрані

# Побудовав гістограму для кожного кольору авто
sns.histplot(
    data=df_colors,
    x='total',
    hue='color',
    multiple='layer',
    palette='tab10',
    hue_order=color_order,
    bins=50,
    alpha=0.6
)

plt.title("Розподіл сум щомісячних платежів по кольорам авто")
plt.xlabel("Сума щомісячних платежів")
plt.ylabel("Кількість поїздок")
plt.tight_layout()
plt.show()
```

12.Дайте відповідь на питання: що означає пряма лінія з точок, паралельна осі, за якою розташована дальність поїздки?

Пряма лінія, паралельна осі X (дальність поїздки), на графіку регресії означає, що для деяких точок лінія тренду не змінюється при зміні дистанції. Тобто, лінія регресії у цих місцях горизонтальна ⇒ немає кореляції між дальністю поїздки та розміром чайових для даного діапазону значень. Часто це трапляється через рядки з нульовою або дуже малою дистанцією, де люди могли платити чайові без поїздки (наприклад, помилки в даних або фіктивні записи). Тобто ці точки "висять" на графіку, але не впливають на нахил загальної лінії тренду сильно.

Щоб наочно побачити вплив нульових або дуже коротких поїздок на лінію тренду чайових, можна відфільтрувати такі поїздки й побудувати графік заново. Наприклад, видалимо всі поїздки з distance == 0 і побачимо, як зміниться регресія.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Відфільтрував дані: тільки поїздки з відстанню більше 0
df_taxi_nonzero = df_taxi_clean[df_taxi_clean['distance'] > 0]

plt.figure(figsize=(10,6))
sns.regplot(
    data=df_taxi_nonzero,
    x='distance',
    y='tip',
    scatter_kws={'alpha':0.5},
    line_kws={'color':'red'}
)
plt.title('Залежність чайових від дальності поїздки (distance>0)')
plt.xlabel('Відстань поїздки')
plt.ylabel('Чайові')
plt.show()
```

Здається, що лінія з суцільних точок паралельна осі X означає, що регресія не бачить жодної залежності між дальністю (distance) та чайовими (tip) у вибірці, яку ми аналізуємо.

- а в поїздках з якого району **найвищий середній розмір** чайових?
- при проїздках з його району розмір чайових найстабільніший (найменше змінюється з дня у день)?

```python
# Видаляю рядки з NA у pickup_borough
df_taxi_clean_no_na = df_taxi_clean[df_taxi_clean['pickup_borough'] != 'NA'].copy()

# Групую по днях та районах і обчислюємо середні tip
borough_tip_pivot = df_taxi_clean_no_na.pivot_table(
    index='pickup_day',
    columns='pickup_borough',
    values='tip',
    aggfunc='mean'
)

# Графік виводжу
borough_tip_pivot.plot(figsize=(12,6), marker='o')
plt.title('Середні денні чайові по районам')
plt.xlabel('Дата поїздки')
plt.ylabel('Середні чайові ($)')
plt.legend(title='Район')
plt.grid(True)
plt.show()
```

2.Середні платежі по типу оплати
Це допоможе зрозуміти, який метод оплати приносить більше доходу та де є ризики.

```python
# Фільтрую NA у payment
df_payment = df_taxi_clean[df_taxi_clean['payment'] != 'NA']

# Середня сума платежів по типу оплати
avg_total_by_payment = df_payment.groupby('payment')['total'].mean().reset_index()

plt.figure(figsize=(8,5))
sns.barplot(x='payment', y='total', data=avg_total_by_payment, palette='pastel')
plt.title('Середня сума платежів по типу оплати')
plt.xlabel('Тип оплати')
plt.ylabel('Середня сума платежів ($)')
plt.show()
```

Метод оплати кредитними картками найприбутковіший.
