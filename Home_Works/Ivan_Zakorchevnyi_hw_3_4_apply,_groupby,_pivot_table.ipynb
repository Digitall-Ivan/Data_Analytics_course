3.1. Створіть зводну таблицю, яка покаже, скільки покупок (інвойсів) було зроблено в кожній з філій (`Branch`) за різними категоріями товарів. Запишіть таблицю в змінну `invoices_by_category` і виведіть змінну на екран.
Ця таблиця допоможе проаналізувати, в якій філії купують найбільше товарів кожної з категорій.

```python
invoices_by_category = df.pivot_table(index='Branch',columns='Product line',values='Invoice ID',aggfunc='count')
print(invoices_by_category)
```

**Питання 1** Яка товарна категорія має найвищий середній чек у кожному місті?
Сенс - спрямувати маркетингові зусилля на продажі даної категорії в кожному місті.
Рекомендація - провести промо акції в кожному місті для категорій товарів з найвищим чеком у найбільш пікові часі (з 12 до 15 та з 19 до 20 годин дня). Додатково дослідити поведінку клієнтів у цих категоріях, провести когортний аналіз клієнтів, виявити найбільш платіжоспроможні групи, спрямувати на них маркетингові зусилля.

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Weekday'] = df['Date'].dt.day_name()
health_beauty_sales = df[df['Product line'] == 'Health and beauty']
health_beauty_sales['Weekday'].value_counts().reset_index()
```

```python
food_beverages_naypyitaw = df[(df['Product line'] == 'Food and beverages') & (df['City'] == 'Naypyitaw')]
food_beverages_naypyitaw['Weekday'].value_counts().reset_index()
```

```python
home_lifestyle_yangon = df[(df['Product line'] == 'Home and lifestyle') & (df['City'] == 'Yangon')]
home_lifestyle_yangon['Weekday'].value_counts().reset_index()
```
